```r
# checking for the missing values
missing_values <- sapply(data, function(x) sum(is.na(x) | x == ''))
missing_values
```

```r
# There are three categorical columns which contains missing values
# Replacing them with most frequent value in respective feature.

# Calculting the mode using fucntion
Mode <- function(x){
    uniqx <- unique(x)
    uniqx[which.max(tabulate(match(x,uniqx)))]
    }

missing_values_columns <- c("Teacher_Quality","Parental_Education_Level","Distance_from_Home")

for (col in missing_values_columns){
    data[[col]] <- ifelse(data[[col]] == '' | is.na(data[[col]]),Mode(data[[col]]),data[[col]])
    }
```

```r
# Removing the outliers using Interquartile range method

remove_outliers <- function(data, column_name){
    Q1 <- quantile(data[[column_name]],0.25,na.rm = TRUE)
    Q3 <- quantile(data[[column_name]],0.75,na.rm = TRUE)
    IQR <- Q3 - Q1
    lower_bound <- Q1 - 1.5 * IQR
    upper_bound <- Q3 + 1.5 * IQR
    data_cleaned <- data[data[[column_name]] >= lower_bound & data[[column_name]] <= upper_bound, ]

    return(data_cleaned)
    }

data <- remove_outliers(data, "Exam_Score")
```

```r
library(MASS)
# Probability thet student scoring exactly 70 based on tutoring sessions
exam_sessions <- data$Tutoring_Sessions[data$Exam_Score == 70]
fit_neg_binom <- fitdistr(exam_sessions, "negative binomial")
print(fit_neg_binom)
```

```r
# Extract parameters
size <- fit_neg_binom$estimate["size"]
mu <- fit_neg_binom$estimate["mu"]

# Generate theoretical probabilities
x_vals <- 0:max(exam_sessions, na.rm = TRUE)
neg_binom_probs <- dnbinom(x_vals, size = size, mu = mu)

observed_table <- as.data.frame(table(factor(exam_sessions, levels = x_vals)))
colnames(observed_table) <- c("Tutoring_Sessions", "Observed_Frequency")
observed_table$Tutoring_Sessions <- as.numeric(as.character(observed_table$Tutoring_Sessions))



observed_table$Observed_Frequency <- observed_table$Observed_Frequency / sum(observed_table$Observed_Frequency)

plot_data <- merge(
  observed_table,
  data.frame(Tutoring_Sessions = x_vals, Theoretical_Probability = neg_binom_probs),
  by = "Tutoring_Sessions",
  all = TRUE
)


plot_data$Observed_Frequency[is.na(plot_data$Observed_Frequency)] <- 0

# Visualization
ggplot(plot_data, aes(x = Tutoring_Sessions)) +
  geom_bar(aes(y = Observed_Frequency), stat = "identity", fill = "skyblue", alpha = 0.6) +
  geom_line(aes(y = Theoretical_Probability), color = "red", size = 1.2) +
  labs(
    title = "Negative Binomial Distribution Fit for Tutoring Sessions",
    x = "Number of Tutoring Sessions",
    y = "Probability / Frequency"
  ) +
  theme_minimal()
```

## 95% Confidence Interval for Exam Score

```r
mean_exam_score <- mean(data$Exam_Score, na.rm = TRUE)
std_error <- sd(data$Exam_Score, na.rm = TRUE) / sqrt(nrow(data))
confidence_interval <- c(mean_exam_score - 1.96 * std_error, mean_exam_score + 1.96 * std_error)



confidence_interval
```

## Confidence Interval for mean exam scores using bootstrapping

```r
# Bootstrapping

library(boot)

# Defining bootstrapping function
bootstrap <- function(data,indices){
    sample_data <- data[indices,]
    mean(sample_data$Exam_Score,na.rm=TRUE)
    }

set.seed(123)
result <- boot(data = data ,statistic=bootstrap,R=1000)
boot.ci(result,type = "perc")
```

```r
# Extrcating bootstrap sample means
boot_means <- result$t

# Extrcating the confidence interval
boot_ci <- boot.ci(result, type = "perc")$percent[4:5]

ggplot(data = data.frame(Bootstrap_Means = boot_means), aes(x = Bootstrap_Means)) +
  geom_density( color = "black", fill = "skyblue", alpha = 0.9) +
  geom_vline(aes(xintercept = mean(boot_means)), color = "red", linetype = "dashed", size = 1.2) +
  geom_vline(aes(xintercept = boot_ci[1]), color = "blue", linetype = "dotted", size = 1.2) +
  geom_vline(aes(xintercept = boot_ci[2]), color = "blue", linetype = "dotted", size = 1.2) +
  labs(
    title = "Bootstrap Distribution of Exam Score Means",
    x = "Bootstrap Sample Means",
    y = "Frequency"
  ) +
  theme_minimal(base_size = 14) +
  theme(plot.title = element_text(hjust = 0.5))
```
